```python
# División del Dataset en datos train y test.
```

```python
archivo=archivo[["PassengerId","Survived","Pclass","Sex","New_Age","SibSp","Parch","Fare"]]
```

```python
X =archivo.drop('Survived', axis=1) # Se cargan todas las columnas excepto la de sobrevivió
Y =archivo['Survived'] # Se cargan solo la columna de sobrevivió
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.25, stratify=Y, random_state=0)
```

```python
print(X_train)
```

```python
print(Y_test)
```

```python
# Entrenamiento del modelo configurado con diferentes hiperparámetros
```

```python
Tarea5-Regresion-Logica = LogisticRegression(solver ="liblinear")
Tarea5-Regresion-Logica.fit (X_train , Y_train)

prediccion = Tarea5-Regresion-Logica.predict(X_test)
```

```python
print(Y_test)
```

```python
print (prediccion)
```

```python
# Evaluación del desempeño del modelo
```

```python
print (classification_report(Y_test,prediccion))
```

```python
# Realizamos las diferentes gráficas que permitiran visualizar los resultados del modelo.
```

```python
cm=confusion_matrix(Y_test,prediccion)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_predictions(Y_test , prediccion)
plt.show()
```

```python
precision = accuracy_score(Y_test, prediccion)
print("Precision:", precision)
```
